```python
import openai
import os
import IPython

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
def get_completion(prompt, model = 'gpt-3.5-turbo'):
    messages=[{"role":"user", "content":prompt}]
    response = openai.ChatCompletion.create(
        model = model,
        messages = messages,
        temperature = 0
    )
    return response.choices[0].message["content"]
```

```python
prompt="""
    You are a web-designer. Your task is to create a website on\
    the popular one-piece anime. You will create the website using\
    html and css. 
"""
response = get_completion(prompt)
print("Website Completed")
print(response)
```

```python
prompt = """You are web designer. Create a web page using HTML \
            and CSS on the popular anime one-piece and write\
            code in your answer. Use headers, footers and div."""
response = get_completion(prompt)
print(response)
```

```python
from IPython.display import display, HTML
```

```python
display(HTML(response))
```
